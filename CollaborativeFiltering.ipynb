```python
moviemat=df1.pivot_table(index='userId',columns='movieId',values='rating')
```

```python
moviemat
```

```python
moviemat.isna().sum()
```

```python
movie1_ratings=moviemat.iloc[:][1]
movie2_ratings=moviemat.iloc[:][2]
```

```python
movie1_similarity=moviemat.corrwith(movie1_ratings,axis=0).sort_values(ascending=False)
```

```python
movie2_similarity=moviemat.corrwith(movie2_ratings,axis=0).sort_values(ascending=False)
```

```python
sim_movies=movie2_similarity.head(20)
```

```python
sim_mov=df1[df1['movieId'].isin(sim_movies)]
```
