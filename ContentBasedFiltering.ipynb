```python
def clean(sen):
    text=re.sub('[^0-9]','',sen)
    return text
```

```python
corpus=[]
ps=PorterStemmer()
ls=WordNetLemmatizer()
```

```python
def cleanData(sen):
    text=re.sub('[^a-zA-Z]',' ',sen)
    text=text.lower()
    text=text.split()
    text=[ls.lemmatize(word) for word in text if not word in stopwords.words('english')]
    sentence=' '.join(text)
    return sentence
```

```python
df1['overview']=df1['overview'].apply(cleanData)
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['name']!=np.nan:
            L.append(i['name'])
    return L
```

```python
df1['genres']=df1['genres'].apply(convert)
```

```python
df1['cast']=df1['cast'].apply(convert)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer(max_features=30000)
```

```python
vectors=tf.fit_transform(movie_df['overview']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie_name):
    idx=movie_df[movie_df['original_title']==movie_name]['id'].values[0]
    similar_movies=sorted(list(enumerate(similarity[idx])),key=lambda x:x[1],reverse=True)

    for movies in similar_movies[0:10]:
        print(movie_df.iloc[movies[0]]['original_title'])
```

```python
recommend("Alien")
```
